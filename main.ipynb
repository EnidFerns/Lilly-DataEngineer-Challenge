## Objectives
1. Calculating the average number of goals per game (1900 and 2000)

```python
df_results_filtered = df_results[(df_results["date"] >= "1900-01-01") & (df_results["date"] < "2000-01-01")].copy()
# df_results_filtered

df_results_filtered["total_goals"] =  df_results_filtered["home_score"] + df_results_filtered["away_score"]
# df_results_filtered

average_goals_per_game = df_results_filtered["total_goals"].mean()
average_goals_per_game
```

```python
with open("answers", "a") as file:
    file.write("\n1. Average number of goals per game between 1900 and 2000:\n")
    file.write(str(average_goals_per_game))
    
print("Average goals per game added to answers.txt")
```

2. Number of shootouts wins by country in alphabetical order

```python
shootout_wins = df_shootouts["winner"].value_counts().reset_index()
shootout_wins.columns = ["country", "shootout_wins"]
shootout_wins = shootout_wins.sort_values("country").reset_index(drop=True)

shootout_wins
```

```python
with open("answers", "a") as file:
    file.write("\n\n2. Number of shootout wins by country in alphabetical order:\n")
    file.write(shootout_wins.to_string(index=False))
    
print("Number of shootout wins by country added to answers.txt")
```

3. key that allows the joining together of goal scorers, results, and shootouts

```python
print(f"df_results shape: {df_results.shape}, df_goalscorers shape: {df_goalscorers.shape}, df_shootouts shape: {df_shootouts.shape}")
```

```python
print(f"df_goalscorers columns: {df_goalscorers.columns}\ndf_results columns: {df_results.columns}\ndf_shootouts columns: {df_shootouts.columns}")
```

```python
# using key as the combination of date, home_team and away_team

df_merged_results = pd.merge(df_results, df_shootouts, on=["date", "home_team", "away_team"], how="left")
df_merged = pd.merge(df_merged_results, df_goalscorers, on=["date", "home_team", "away_team"], how="left")
# df_merged_results
df_merged
```

4. Teams that have won a penalty shootout after a 1-1 draw

```python
df_merged["penalty"].unique()
```

```python
# only combination of shootouts and results required
df_draws_1_1 = df_merged[(df_merged["home_score"] == 1) & (df_merged["away_score"] == 1) & (df_merged["penalty"] == True)]
# df_draws_1_1
shootout_winners = df_draws_1_1[["tournament","home_team","away_team","winner"]]
# shootout_winners = df_draws_1_1[["winner"]]
shootout_winners = shootout_winners.dropna()
shootout_winners = shootout_winners.reset_index(drop=True)
shootout_winners
```

```python
with open("answers", "a") as file:
    file.write("\n\n4. Teams that have won a penalty shootout after a 1-1 draw:\n")
    file.write(shootout_winners.to_string(index=False))
    
print("Teams that won a shootout after a 1-1 draw added to answers.txt")
```

5. Top goal scorers by tournament, and what percentage that equates to for all goals scored in the tournament.

```python
df_merged["total_goals"] =  df_merged["home_score"] + df_merged["away_score"]
df_top_scorer = df_merged[["tournament","home_team","away_team","total_goals","scorer"]].copy()
# df_top_scorer

goals_per_player = df_top_scorer.groupby(["tournament", "scorer"]).size().reset_index()
goals_per_player.columns = ["tournament", "scorer","goals_scored"]
goals_per_player = goals_per_player.sort_values("goals_scored",ascending=False)
# goals_per_player

top_scorers = goals_per_player.groupby("tournament").head(1)
# top_scorers

total_goals = df_top_scorer.groupby("tournament")["total_goals"].sum()
total_goals.columns = ["tournament", "total_goals"]
# total_goals

df_top_scorers = pd.merge(total_goals, top_scorers, on=["tournament"], how='inner')
df_top_scorers["contribution_percentage"] = (df_top_scorers["goals_scored"] / df_top_scorers["total_goals"]) * 100
df_top_scorers_with_percentage = df_top_scorers[["tournament","scorer","contribution_percentage"]].copy()
df_top_scorers_with_percentage
```

```python
with open("answers", "a") as file:
    file.write("\n\n5. Top goal scorers by tournament, along with percentage that equates to for all goals scored in the tournament:\n")
    file.write(df_top_scorers_with_percentage.to_string(index=False))
    
print("Top scorers of each tournament along with the percentage of all goals scored added to answers.txt")
```
